```python
#!/usr/bin/env python3

PHASE 5: XGBOOST + SHAP ANALYSIS (CORRECTED)
Project: hormonesbp


import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (accuracy_score, roc_auc_score, matthews_corrcoef, 
                             confusion_matrix, classification_report)
from xgboost import XGBClassifier
import shap
import matplotlib.pyplot as plt
import seaborn as sns

# Set working directory
os.chdir("/Users/medhajha/Downloads/paper/hormonesbrpr/hormonesbp")

print("\n" + "="*80)
print("PHASE 5: XGBOOST + SHAP ANALYSIS")
print("Working Directory:", os.getcwd())
print("="*80 + "\n")


# ANALYSIS 1: TUMOR VS NORMAL

print("="*80)
print("ANALYSIS 1: TUMOR VS NORMAL CLASSIFICATION")
print("="*80)

try:
    print("\n>>> Loading Tumor vs Normal data...")
    dat1 = pd.read_csv("results/data/expression_tumor_vs_normal.csv")
    
    metadata_cols = ['label', 'cancerType', 'sampleID', 'sampleType']
    gene_cols = [col for col in dat1.columns if col not in metadata_cols]
    
    X1 = dat1[gene_cols]
    y1 = dat1['label'].values
    
    print(f"✓ Data loaded: {X1.shape}")
    print(f"  Tumor: {sum(y1 == 1)}, Normal: {sum(y1 == 0)}")
    print(f"  Genes: {X1.shape[1]}")
    
    # Split
    X1_train, X1_test, y1_train, y1_test = train_test_split(
        X1, y1, test_size=0.2, random_state=42, stratify=y1
    )
    print(f"\n>>> Split: Train {X1_train.shape[0]}, Test {X1_test.shape[0]}")
    
    # Train XGBoost
    print("\n>>> Training XGBoost...")
    xgb1 = XGBClassifier(
        n_estimators=100, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        random_state=42, use_label_encoder=False,
        eval_metric="logloss", verbose=0
    )
    
    # CV
    print(">>> 5-fold cross-validation...")
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_acc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="accuracy")
    cv_auc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="roc_auc")
    cv_mcc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="matthews_corrcoef")
    
    print(f"\n✓ CV Results (Tumor vs Normal):")
    print(f"  Accuracy: {cv_acc.mean():.4f} ± {cv_acc.std():.4f}")
    print(f"  AUC-ROC:  {cv_auc.mean():.4f} ± {cv_auc.std():.4f}")
    print(f"  MCC:      {cv_mcc.mean():.4f} ± {cv_mcc.std():.4f}")
    
    # Fit
    xgb1.fit(X1_train, y1_train)
    
    # Test
    print("\n>>> Test Set:")
    y1_pred = xgb1.predict(X1_test)
    y1_prob = xgb1.predict_proba(X1_test)[:, 1]
    
    test_acc = accuracy_score(y1_test, y1_pred)
    test_auc = roc_auc_score(y1_test, y1_prob)
    test_mcc = matthews_corrcoef(y1_test, y1_pred)
    
    print(f"  Accuracy: {test_acc:.4f}")
    print(f"  AUC-ROC:  {test_auc:.4f}")
    print(f"  MCC:      {test_mcc:.4f}")
    
    cm1 = confusion_matrix(y1_test, y1_pred)
    sensitivity1 = cm1[1,1] / (cm1[1,1] + cm1[1,0]) if (cm1[1,1] + cm1[1,0]) > 0 else 0
    specificity1 = cm1[0,0] / (cm1[0,0] + cm1[0,1]) if (cm1[0,0] + cm1[0,1]) > 0 else 0
    
    print(f"  Sensitivity: {sensitivity1:.4f}, Specificity: {specificity1:.4f}")
    
    # SHAP
    print("\n>>> SHAP Analysis...")
    explainer1 = shap.TreeExplainer(xgb1)
    shap_values1 = explainer1.shap_values(X1_test)
    
    print("  Creating plots...")
    plt.figure(figsize=(14, 8))
    shap.summary_plot(shap_values1, X1_test, show=False, plot_type="violin", max_display=20)
    plt.title("SHAP Summary: Tumor vs Normal", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure7_SHAP_summary_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("   Figure7_SHAP_summary_tumor_vs_normal.png")
    
    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_values1, X1_test, show=False, plot_type="bar", max_display=15)
    plt.title("SHAP Feature Importance: Tumor vs Normal", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure8_SHAP_bar_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("   Figure8_SHAP_bar_tumor_vs_normal.png")
    
    # Top genes
    mean_shap1 = np.abs(shap_values1).mean(axis=0)
    top_genes1 = pd.DataFrame({
        "gene": X1.columns,
        "mean_abs_shap": mean_shap1
    }).sort_values("mean_abs_shap", ascending=False)
    
    top_genes1.to_csv("results/tables/Table5_top_genes_SHAP_tumor_vs_normal.csv", index=False)
    
    print(f"\n✓ Top genes (Tumor vs Normal):")
    print(top_genes1.head(10).to_string(index=False))
    
except Exception as e:
    print(f"\n ERROR in Analysis 1: {str(e)}")
    import traceback
    traceback.print_exc()


# ANALYSIS 2: BRCA VS PRAD

print("\n\n" + "="*80)
print("ANALYSIS 2: BRCA VS PRAD CLASSIFICATION")
print("="*80)

try:
    print("\n>>> Loading BRCA vs PRAD data...")
    dat2 = pd.read_csv("results/data/expression_brca_vs_prad.csv")
    
    metadata_cols2 = ['label', 'tumorStatus', 'sampleID']
    gene_cols2 = [col for col in dat2.columns if col not in metadata_cols2]
    
    X2 = dat2[gene_cols2]
    y2 = dat2['label'].values
    
    print(f"✓ Data loaded: {X2.shape}")
    print(f"  BRCA: {sum(y2 == 1)}, PRAD: {sum(y2 == 0)}")
    print(f"  Genes: {X2.shape[1]}")
    
    # Split
    X2_train, X2_test, y2_train, y2_test = train_test_split(
        X2, y2, test_size=0.2, random_state=42, stratify=y2
    )
    print(f"\n>>> Split: Train {X2_train.shape[0]}, Test {X2_test.shape[0]}")
    
    # Train XGBoost
    print("\n>>> Training XGBoost...")
    xgb2 = XGBClassifier(
        n_estimators=100, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        random_state=42, use_label_encoder=False,
        eval_metric="logloss", verbose=0
    )
    
    # CV
    print(">>> 5-fold cross-validation...")
    cv_acc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="accuracy")
    cv_auc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="roc_auc")
    cv_mcc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="matthews_corrcoef")
    
    print(f"\n✓ CV Results (BRCA vs PRAD):")
    print(f"  Accuracy: {cv_acc2.mean():.4f} ± {cv_acc2.std():.4f}")
    print(f"  AUC-ROC:  {cv_auc2.mean():.4f} ± {cv_auc2.std():.4f}")
    print(f"  MCC:      {cv_mcc2.mean():.4f} ± {cv_mcc2.std():.4f}")
    
    # Fit
    xgb2.fit(X2_train, y2_train)
    
    # Test
    print("\n>>> Test Set:")
    y2_pred = xgb2.predict(X2_test)
    y2_prob = xgb2.predict_proba(X2_test)[:, 1]
    
    test_acc2 = accuracy_score(y2_test, y2_pred)
    test_auc2 = roc_auc_score(y2_test, y2_prob)
    test_mcc2 = matthews_corrcoef(y2_test, y2_pred)
    
    print(f"  Accuracy: {test_acc2:.4f}")
    print(f"  AUC-ROC:  {test_auc2:.4f}")
    print(f"  MCC:      {test_mcc2:.4f}")
    
    cm2 = confusion_matrix(y2_test, y2_pred)
    sensitivity2 = cm2[1,1] / (cm2[1,1] + cm2[1,0]) if (cm2[1,1] + cm2[1,0]) > 0 else 0
    specificity2 = cm2[0,0] / (cm2[0,0] + cm2[0,1]) if (cm2[0,0] + cm2[0,1]) > 0 else 0
    
    print(f"  Sensitivity: {sensitivity2:.4f}, Specificity: {specificity2:.4f}")
    
    # SHAP
    print("\n>>> SHAP Analysis...")
    explainer2 = shap.TreeExplainer(xgb2)
    shap_values2 = explainer2.shap_values(X2_test)
    
    print("  Creating plots...")
    plt.figure(figsize=(14, 8))
    shap.summary_plot(shap_values2, X2_test, show=False, plot_type="violin", max_display=15)
    plt.title("SHAP Summary: BRCA vs PRAD", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure9_SHAP_summary_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("   Figure9_SHAP_summary_brca_vs_prad.png")
    
    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_values2, X2_test, show=False, plot_type="bar", max_display=15)
    plt.title("SHAP Feature Importance: BRCA vs PRAD", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure10_SHAP_bar_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("   Figure10_SHAP_bar_brca_vs_prad.png")
    
    # Top genes
    mean_shap2 = np.abs(shap_values2).mean(axis=0)
    top_genes2 = pd.DataFrame({
        "gene": X2.columns,
        "mean_abs_shap": mean_shap2
    }).sort_values("mean_abs_shap", ascending=False)
    
    top_genes2.to_csv("results/tables/Table6_top_genes_SHAP_brca_vs_prad.csv", index=False)
    
    print(f"\n✓ Top genes (BRCA vs PRAD):")
    print(top_genes2.head(10).to_string(index=False))
    
except Exception as e:
    print(f"\n ERROR in Analysis 2: {str(e)}")
    import traceback
    traceback.print_exc()


# SUMMARY

try:
    print("\n>>> Creating summary table...")
    
    summary_df = pd.DataFrame({
        'Analysis': ['Tumor vs Normal', 'BRCA vs PRAD'],
        'CV_Accuracy': [f"{cv_acc.mean():.4f}", f"{cv_acc2.mean():.4f}"],
        'Test_Accuracy': [f"{test_acc:.4f}", f"{test_acc2:.4f}"],
        'Test_AUC': [f"{test_auc:.4f}", f"{test_auc2:.4f}"]
    })
    
    summary_df.to_csv("results/tables/Table7_model_performance_summary.csv", index=False)
    
    print("\n" + "="*80)
    print("PHASE 5 COMPLETE: XGBOOST + SHAP ANALYSIS ")
    print("="*80)
    
    print("\n✓ PERFORMANCE SUMMARY:\n")
    print(summary_df.to_string(index=False))
    
    print("\n✓ FILES CREATED:")
    print("  Figures: Figure7, Figure8, Figure9, Figure10")
    print("  Tables: Table5, Table6, Table7")
    
    print("\n" + "="*80)
    print("READY FOR PHASE 6: SURVIVAL ANALYSIS")
    print("="*80 + "\n")
    
except Exception as e:
    print(f"\n ERROR in summary: {str(e)}")
```

```python
#!/usr/bin/env python3
"""
==============================================================================
PHASE 5: XGBOOST + SHAP ANALYSIS 
Project: hormonesbp
==============================================================================
"""

import os
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.metrics import (accuracy_score, roc_auc_score, matthews_corrcoef, 
                             confusion_matrix, classification_report, roc_curve, auc)
from xgboost import XGBClassifier
import shap
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# Set working directory
os.chdir("/Users/medhajha/Downloads/paper/hormonesbrpr/hormonesbp")

print("\n" + "="*80)
print("PHASE 5: XGBOOST + SHAP ANALYSIS (Publication Quality)")
print("Working Directory:", os.getcwd())
print("="*80 + "\n")

# ==============================================================================
# ANALYSIS 1: TUMOR VS NORMAL
# ==============================================================================

print("="*80)
print("ANALYSIS 1: TUMOR VS NORMAL CLASSIFICATION")
print("="*80)

try:
    print("\n>>> Loading Tumor vs Normal data...")
    dat1 = pd.read_csv("results/data/expression_tumor_vs_normal.csv")
    
    metadata_cols = ['label', 'cancerType', 'sampleID', 'sampleType']
    gene_cols = [col for col in dat1.columns if col not in metadata_cols]
    
    X1 = dat1[gene_cols]
    y1 = dat1['label'].values
    
    print(f"✓ Data loaded: {X1.shape}")
    print(f"  Tumor: {sum(y1 == 1)}, Normal: {sum(y1 == 0)}")
    
    # Split
    X1_train, X1_test, y1_train, y1_test = train_test_split(
        X1, y1, test_size=0.2, random_state=42, stratify=y1
    )
    print(f"✓ Train: {X1_train.shape[0]}, Test: {X1_test.shape[0]}")
    
    # Train XGBoost
    print("\n>>> Training XGBoost...")
    xgb1 = XGBClassifier(
        n_estimators=100, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        random_state=42, use_label_encoder=False,
        eval_metric="logloss", verbose=0
    )
    
    # CV
    print(">>> 5-fold cross-validation...")
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_acc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="accuracy")
    cv_auc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="roc_auc")
    cv_mcc = cross_val_score(xgb1, X1_train, y1_train, cv=cv, scoring="matthews_corrcoef")
    
    print(f"\n✓ 5-Fold CV Results (Tumor vs Normal):")
    print(f"  Accuracy: {cv_acc.mean():.4f} ± {cv_acc.std():.4f}")
    print(f"  AUC-ROC:  {cv_auc.mean():.4f} ± {cv_auc.std():.4f}")
    print(f"  MCC:      {cv_mcc.mean():.4f} ± {cv_mcc.std():.4f}")
    
    # Fit
    xgb1.fit(X1_train, y1_train)
    
    # Test
    print("\n>>> Test Set Results:")
    y1_pred = xgb1.predict(X1_test)
    y1_prob = xgb1.predict_proba(X1_test)[:, 1]
    
    test_acc = accuracy_score(y1_test, y1_pred)
    test_auc = roc_auc_score(y1_test, y1_prob)
    test_mcc = matthews_corrcoef(y1_test, y1_pred)
    
    print(f"  Accuracy: {test_acc:.4f}")
    print(f"  AUC-ROC:  {test_auc:.4f}")
    print(f"  MCC:      {test_mcc:.4f}")
    
    cm1 = confusion_matrix(y1_test, y1_pred)
    sensitivity1 = cm1[1,1] / (cm1[1,1] + cm1[1,0])
    specificity1 = cm1[0,0] / (cm1[0,0] + cm1[0,1])
    
    print(f"\n  Confusion Matrix:")
    print(f"    TN: {cm1[0,0]}, FP: {cm1[0,1]}")
    print(f"    FN: {cm1[1,0]}, TP: {cm1[1,1]}")
    print(f"  Sensitivity (TPR): {sensitivity1:.4f}")
    print(f"  Specificity (TNR): {specificity1:.4f}")
    
    # ========== VISUALIZATION 1: Confusion Matrix Heatmap ==========
    print("\n>>> Creating Confusion Matrix heatmap...")
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Normal', 'Tumor'],
                yticklabels=['Normal', 'Tumor'],
                ax=ax, annot_kws={'size': 14})
    ax.set_ylabel('True Label', fontsize=12, fontweight='bold')
    ax.set_xlabel('Predicted Label', fontsize=12, fontweight='bold')
    ax.set_title('Confusion Matrix: Tumor vs Normal\nAccuracy={:.4f}, AUC-ROC={:.4f}'.format(test_acc, test_auc),
                fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.savefig("results/figures/Figure7A_confusion_matrix_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure7A_confusion_matrix_tumor_vs_normal.png")
    
    # ========== VISUALIZATION 2: ROC Curve ==========
    print(">>> Creating ROC curve...")
    fpr1, tpr1, _ = roc_curve(y1_test, y1_prob)
    roc_auc1 = auc(fpr1, tpr1)
    
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.plot(fpr1, tpr1, color='darkorange', lw=3, label=f'ROC Curve (AUC = {roc_auc1:.4f})')
    ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
    ax.fill_between(fpr1, tpr1, alpha=0.2, color='darkorange')
    ax.set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
    ax.set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
    ax.set_title('ROC Curve: Tumor vs Normal Classification', fontsize=14, fontweight='bold', pad=20)
    ax.legend(loc="lower right", fontsize=11)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig("results/figures/Figure7B_roc_curve_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure7B_roc_curve_tumor_vs_normal.png")
    
    # ========== SHAP ANALYSIS ==========
    print("\n>>> SHAP Analysis...")
    explainer1 = shap.TreeExplainer(xgb1)
    shap_values1 = explainer1.shap_values(X1_test)
    
    # SHAP Summary Plot (Violin - like Datta)
    print("  Creating SHAP summary plot (violin)...")
    fig, ax = plt.subplots(figsize=(14, 9))
    shap.summary_plot(shap_values1, X1_test, show=False, plot_type="violin", max_display=20)
    plt.title("SHAP Summary Plot: Tumor vs Normal\n(Violin plot showing feature distributions)", 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel("SHAP Value (Model Output Impact)", fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure7C_SHAP_summary_violin_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure7C_SHAP_summary_violin_tumor_vs_normal.png")
    
    # SHAP Bar Plot (Feature Importance)
    print("  Creating SHAP bar plot...")
    fig, ax = plt.subplots(figsize=(12, 8))
    shap.summary_plot(shap_values1, X1_test, show=False, plot_type="bar", max_display=15)
    plt.title("SHAP Feature Importance: Tumor vs Normal\n(Mean |SHAP| values)", 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel("Mean |SHAP Value|", fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure7D_SHAP_bar_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure7D_SHAP_bar_tumor_vs_normal.png")
    
    # SHAP Decision Plot (like Datta's)
    print("  Creating SHAP decision plot...")
    try:
        fig, ax = plt.subplots(figsize=(14, 8))
        shap.decision_plot(explainer1.expected_value, shap_values1[:50], X1_test.iloc[:50], 
                          highlight=y1_test[:50], color_bar=False, show=False)
        plt.title("SHAP Decision Plot: Tumor vs Normal\n(First 50 test samples)", 
                  fontsize=14, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.savefig("results/figures/Figure7E_SHAP_decision_tumor_vs_normal.png", dpi=300, bbox_inches='tight')
        plt.close()
        print("  ✓ Figure7E_SHAP_decision_tumor_vs_normal.png")
    except Exception as e:
        print(f"   Decision plot skipped: {str(e)}")
    
    # Top genes
    mean_shap1 = np.abs(shap_values1).mean(axis=0)
    top_genes1 = pd.DataFrame({
        "gene": X1.columns,
        "mean_abs_shap": mean_shap1
    }).sort_values("mean_abs_shap", ascending=False)
    
    top_genes1.to_csv("results/tables/Table5_top_genes_SHAP_tumor_vs_normal.csv", index=False)
    
    print(f"\n✓ Top 10 genes (Tumor vs Normal):")
    print(top_genes1.head(10).to_string(index=False))
    
except Exception as e:
    print(f"\n ERROR in Analysis 1: {str(e)}")
    import traceback
    traceback.print_exc()

# ==============================================================================
# ANALYSIS 2: BRCA VS PRAD
# ==============================================================================

print("\n\n" + "="*80)
print("ANALYSIS 2: BRCA VS PRAD CLASSIFICATION")
print("="*80)

try:
    print("\n>>> Loading BRCA vs PRAD data...")
    dat2 = pd.read_csv("results/data/expression_brca_vs_prad.csv")
    
    metadata_cols2 = ['label', 'tumorStatus', 'sampleID']
    gene_cols2 = [col for col in dat2.columns if col not in metadata_cols2]
    
    X2 = dat2[gene_cols2]
    y2 = dat2['label'].values
    
    print(f"✓ Data loaded: {X2.shape}")
    print(f"  BRCA: {sum(y2 == 1)}, PRAD: {sum(y2 == 0)}")
    
    # Split
    X2_train, X2_test, y2_train, y2_test = train_test_split(
        X2, y2, test_size=0.2, random_state=42, stratify=y2
    )
    print(f"✓ Train: {X2_train.shape[0]}, Test: {X2_test.shape[0]}")
    
    # Train XGBoost
    print("\n>>> Training XGBoost...")
    xgb2 = XGBClassifier(
        n_estimators=100, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        random_state=42, use_label_encoder=False,
        eval_metric="logloss", verbose=0
    )
    
    # CV
    print(">>> 5-fold cross-validation...")
    cv_acc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="accuracy")
    cv_auc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="roc_auc")
    cv_mcc2 = cross_val_score(xgb2, X2_train, y2_train, cv=cv, scoring="matthews_corrcoef")
    
    print(f"\n✓ 5-Fold CV Results (BRCA vs PRAD):")
    print(f"  Accuracy: {cv_acc2.mean():.4f} ± {cv_acc2.std():.4f}")
    print(f"  AUC-ROC:  {cv_auc2.mean():.4f} ± {cv_auc2.std():.4f}")
    print(f"  MCC:      {cv_mcc2.mean():.4f} ± {cv_mcc2.std():.4f}")
    
    # Fit
    xgb2.fit(X2_train, y2_train)
    
    # Test
    print("\n>>> Test Set Results:")
    y2_pred = xgb2.predict(X2_test)
    y2_prob = xgb2.predict_proba(X2_test)[:, 1]
    
    test_acc2 = accuracy_score(y2_test, y2_pred)
    test_auc2 = roc_auc_score(y2_test, y2_prob)
    test_mcc2 = matthews_corrcoef(y2_test, y2_pred)
    
    print(f"  Accuracy: {test_acc2:.4f}")
    print(f"  AUC-ROC:  {test_auc2:.4f}")
    print(f"  MCC:      {test_mcc2:.4f}")
    
    cm2 = confusion_matrix(y2_test, y2_pred)
    sensitivity2 = cm2[1,1] / (cm2[1,1] + cm2[1,0])
    specificity2 = cm2[0,0] / (cm2[0,0] + cm2[0,1])
    
    print(f"\n  Confusion Matrix:")
    print(f"    PRAD: {cm2[0,0]}, Misclassified as BRCA: {cm2[0,1]}")
    print(f"    Misclassified as PRAD: {cm2[1,0]}, BRCA: {cm2[1,1]}")
    print(f"  Sensitivity (BRCA): {sensitivity2:.4f}")
    print(f"  Specificity (PRAD): {specificity2:.4f}")
    
    # ========== VISUALIZATION 1: Confusion Matrix Heatmap ==========
    print("\n>>> Creating Confusion Matrix heatmap...")
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(cm2, annot=True, fmt='d', cmap='Greens', cbar=False,
                xticklabels=['PRAD', 'BRCA'],
                yticklabels=['PRAD', 'BRCA'],
                ax=ax, annot_kws={'size': 14})
    ax.set_ylabel('True Label', fontsize=12, fontweight='bold')
    ax.set_xlabel('Predicted Label', fontsize=12, fontweight='bold')
    ax.set_title('Confusion Matrix: BRCA vs PRAD\nAccuracy={:.4f}, AUC-ROC={:.4f}'.format(test_acc2, test_auc2),
                fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.savefig("results/figures/Figure8A_confusion_matrix_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure8A_confusion_matrix_brca_vs_prad.png")
    
    # ========== VISUALIZATION 2: ROC Curve ==========
    print(">>> Creating ROC curve...")
    fpr2, tpr2, _ = roc_curve(y2_test, y2_prob)
    roc_auc2 = auc(fpr2, tpr2)
    
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.plot(fpr2, tpr2, color='darkgreen', lw=3, label=f'ROC Curve (AUC = {roc_auc2:.4f})')
    ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
    ax.fill_between(fpr2, tpr2, alpha=0.2, color='darkgreen')
    ax.set_xlabel('False Positive Rate', fontsize=12, fontweight='bold')
    ax.set_ylabel('True Positive Rate', fontsize=12, fontweight='bold')
    ax.set_title('ROC Curve: BRCA vs PRAD Classification', fontsize=14, fontweight='bold', pad=20)
    ax.legend(loc="lower right", fontsize=11)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig("results/figures/Figure8B_roc_curve_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure8B_roc_curve_brca_vs_prad.png")
    
    # ========== SHAP ANALYSIS ==========
    print("\n>>> SHAP Analysis...")
    explainer2 = shap.TreeExplainer(xgb2)
    shap_values2 = explainer2.shap_values(X2_test)
    
    # SHAP Summary Plot (Violin)
    print("  Creating SHAP summary plot (violin)...")
    fig, ax = plt.subplots(figsize=(14, 9))
    shap.summary_plot(shap_values2, X2_test, show=False, plot_type="violin", max_display=15)
    plt.title("SHAP Summary Plot: BRCA vs PRAD\n(Violin plot showing feature distributions)", 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel("SHAP Value (Model Output Impact)", fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure8C_SHAP_summary_violin_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure8C_SHAP_summary_violin_brca_vs_prad.png")
    
    # SHAP Bar Plot
    print("  Creating SHAP bar plot...")
    fig, ax = plt.subplots(figsize=(12, 8))
    shap.summary_plot(shap_values2, X2_test, show=False, plot_type="bar", max_display=15)
    plt.title("SHAP Feature Importance: BRCA vs PRAD\n(Mean |SHAP| values)", 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel("Mean |SHAP Value|", fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig("results/figures/Figure8D_SHAP_bar_brca_vs_prad.png", dpi=300, bbox_inches='tight')
    plt.close()
    print("  ✓ Figure8D_SHAP_bar_brca_vs_prad.png")
    
    # SHAP Decision Plot
    print("  Creating SHAP decision plot...")
    try:
        fig, ax = plt.subplots(figsize=(14, 8))
        shap.decision_plot(explainer2.expected_value, shap_values2[:50], X2_test.iloc[:50], 
                          highlight=y2_test[:50], color_bar=False, show=False)
        plt.title("SHAP Decision Plot: BRCA vs PRAD\n(First 50 test samples)", 
                  fontsize=14, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.savefig("results/figures/Figure8E_SHAP_decision_brca_vs_prad.png", dpi=300, bbox_inches='tight')
        plt.close()
        print("  ✓ Figure8E_SHAP_decision_brca_vs_prad.png")
    except Exception as e:
        print(f"   Decision plot skipped: {str(e)}")
    
    # Top genes
    mean_shap2 = np.abs(shap_values2).mean(axis=0)
    top_genes2 = pd.DataFrame({
        "gene": X2.columns,
        "mean_abs_shap": mean_shap2
    }).sort_values("mean_abs_shap", ascending=False)
    
    top_genes2.to_csv("results/tables/Table6_top_genes_SHAP_brca_vs_prad.csv", index=False)
    
    print(f"\n✓ Top 10 genes (BRCA vs PRAD):")
    print(top_genes2.head(10).to_string(index=False))
    
except Exception as e:
    print(f"\n ERROR in Analysis 2: {str(e)}")
    import traceback
    traceback.print_exc()

# ==============================================================================
# SUMMARY
# ==============================================================================

try:
    print("\n>>> Creating summary table...")
    
    summary_df = pd.DataFrame({
        'Analysis': ['Tumor vs Normal', 'BRCA vs PRAD'],
        'CV_Accuracy': [f"{cv_acc.mean():.4f} ± {cv_acc.std():.4f}", 
                       f"{cv_acc2.mean():.4f} ± {cv_acc2.std():.4f}"],
        'CV_AUC': [f"{cv_auc.mean():.4f} ± {cv_auc.std():.4f}",
                  f"{cv_auc2.mean():.4f} ± {cv_auc2.std():.4f}"],
        'Test_Accuracy': [f"{test_acc:.4f}", f"{test_acc2:.4f}"],
        'Test_AUC': [f"{test_auc:.4f}", f"{test_auc2:.4f}"],
        'Test_MCC': [f"{test_mcc:.4f}", f"{test_mcc2:.4f}"]
    })
    
    summary_df.to_csv("results/tables/Table7_model_performance_summary.csv", index=False)
    
    print("\n" + "="*80)
    print("PHASE 5 COMPLETE: XGBOOST + SHAP ANALYSIS ✓")
    print("="*80)
    
    print("\n✓ PERFORMANCE SUMMARY:\n")
    print(summary_df.to_string(index=False))
    
    print("\n\n✓ FILES CREATED (Publication Quality - Datta Style):\n")
    print("  Analysis 1 - Tumor vs Normal:")
    print("    ✓ Figure7A: Confusion Matrix")
    print("    ✓ Figure7B: ROC Curve")
    print("    ✓ Figure7C: SHAP Violin Plot")
    print("    ✓ Figure7D: SHAP Bar Plot")
    print("    ✓ Figure7E: SHAP Decision Plot")
    print("    ✓ Table5: Top genes by SHAP")
    print("\n  Analysis 2 - BRCA vs PRAD:")
    print("    ✓ Figure8A: Confusion Matrix")
    print("    ✓ Figure8B: ROC Curve")
    print("    ✓ Figure8C: SHAP Violin Plot")
    print("    ✓ Figure8D: SHAP Bar Plot")
    print("    ✓ Figure8E: SHAP Decision Plot")
    print("    ✓ Table6: Top genes by SHAP")
    print("\n  Summary:")
    print("    ✓ Table7: Model Performance Summary")
    
    print("\n" + "="*80)
    print("READY FOR PHASE 6: SURVIVAL ANALYSIS & PATHWAY ENRICHMENT")
    print("="*80 + "\n")
    
except Exception as e:
    print(f"\n ERROR in summary: {str(e)}")
    import traceback
    traceback.print_exc()
```
